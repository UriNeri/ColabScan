## Enviroment setup
First, we need to install the dependencies required.


```python
%%shell
wget https://github.com/shenwei356/seqkit/releases/download/v2.4.0/seqkit_linux_amd64.tar.gz --quiet
tar -zxvf  seqkit_linux_amd64.tar.gz

wget http://eddylab.org/software/hmmer/hmmer.tar.gz --quiet
tar -zxvf hmmer.tar.gz
cd hmmer-3.3.2
./configure
make
make install
# workaround for permission trouble or if you are running locally, aliasing to local path
alias hmmbuild='$(pwd)/src/hmmbuild'
alias hmmsearch='$(pwd)/src/hmmsearch'
cd ..
alias seqkit='$(pwd)/seqkit'

```

## Data fetching
Next we're gonna fetch several public resources we'll search your data agains (RdRp-scan, NeoRdRp, RVMT, and hopefully more in the future).


```python
%%shell
wget https://github.com/shoichisakaguchi/NeoRdRp/raw/main/NeoRdRp-HMM.v1.1.hmm.gz  --quiet
gzip -d NeoRdRp-HMM.v1.1.hmm.gz

git clone --recursive https://github.com/JustineCharon/RdRp-scan

## wget https://zenodo.org/record/7368133/files/RVMT_Zenodo_V4.zip?download=1 
# *too big, download would take 4 hours, going to improvise:

wget https://portal.nersc.gov/dna/microbial/prokpubs/Riboviria/RiboV1.4/Alignments/zip.ali.220515.tgz  --quiet
tar -zxvf zip.ali.220515.tgz
mkdir RVMT_HMMs

for ali_folder in ./ali.*
do
cd $ali_folder
for alignm_file in ./*.FASTA
do
ali_name=$(basename $alignm_file .FASTA)
hmmbuild  -n "$ali_name"_RVMT  -o /dev/null ../RVMT_HMMs/"$ali_name".hmm $alignm_file
done
cd ..
done

##### Concatenate the HMM (HMMER) profiles into a single hmmDB. #####
cd RVMT_HMMs
cat ./*hmm  > ../RVMT.hmm
cd ..
```

## HMMsearching  
Next we are going to do start an hmmsearch against the different

```python
%%shell
cd /content/

HMM_DBs=(/content/RVMT.hmm /content/NeoRdRp-HMM.v1.1.hmm /content/RdRp-scan/Profile_db_and_alignments/RdRp_HMM_profile_CLUSTALO.db)
for hmmdb in ${HMM_DBs[*]}; do
  hmmdb_name=$(basename $hmmdb .hmm)
intput_file=./*.fasta
ini_name=$(basename $intput_file ".fasta")
echo $ini_name
echo $eval
echo $score
./seqkit translate -x -F --clean -w 20000 -f 6 $intput_file --id-regexp "(\*)" > ./"$ini_name".faa
hmmsearch  -E $eval --domE $eval --incdomE $eval --incE $eval -o "$hmmdb_name"_hmmsearch_vs_"$ini_name".tsv --domtblout domtblout_"$hmmdb_name"_hmmsearch_vs_"$ini_name".tsv --tblout tabular_"$hmmdb_name"_hmmsearch_vs_"$ini_name".tsv $hmmdb ./"$ini_name".faa
sed -i '/^#/d' domtblout_"$hmmdb_name"_hmmsearch_vs_"$ini_name".tsv
# write the headers
echo "target_name	target_accession	qL	query_name	query_accession	pL	E-value_fullseq	score_fullseq	bias_fullseq	#	of	c-Evalue	i-Evalue	bias_thisdomain	score_thisdomain	p1	p2	q1	q2	env_from	env_to	acc	description_of_target" >awked_domtblout_"$hmmdb_name"_hmmsearch_vs_"$ini_name".tsv
# organize the output neat-ly-er
awk '{print $1"\t"$2"\t"$3"\t"$4"\t"$5"\t"$6"\t"$7"\t"$8"\t"$9"\t"$10"\t"$11"\t"$12"\t"$13"\t"$14"\t"$15"\t"$16"\t"$17"\t"$18"\t"$19"\t"$20"\t"$21"\t"$22"\t"$23"\t"$24}' domtblout_"$hmmdb_name"_hmmsearch_vs_"$ini_name".tsv >>awked_domtblout_"$hmmdb_name"_hmmsearch_vs_"$ini_name".tsv #Legacy - don't chage to OFS because of trailing sep issue with parser
done
```
